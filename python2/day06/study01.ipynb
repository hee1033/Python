```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm 
import matplotlib

fig = plt.figure(constrained_layout =True)
font_location = "C:\Windows\Fonts\H2GTRE.TTF"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font',family = fontname)

# 막대 그래프
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)

x = [3,4]
y = [1,2]

labels = ['english','math']
colors = ['red','blue']
ax.bar(x,y, tick_label = labels, color = colors)

for i, j  in enumerate(y):
    ax.text(j+2, i+1.05,str(j),color = 'green',fontweight = 'bold')
plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english','math']
x = [3,4]
y = [1,2]
colors = ['red','blue']
ax.barh(x,y,tick_label = labels,color = colors)
plt.show()
```

```python
import numpy as np
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)

x = [1,2]
y1,y2,y3 = [1,2],[2,5],[3,7]
w = 0.2
ax.bar(x,y1,width=w,label = 'english')
ax.bar(np.array(x)+w,y2,width=w,label = 'math')
ax.bar(np.array(x)+w+w,y3,width=w,label = 'korean')

plt.show()
```

```python
# pie 차트 만들기
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['english','math','korean']
sizes = [45,35,20]
ax.pie(sizes, labels = label,autopct='%1.1f%%',shadow= True)

# 차트 제목 설정
ax.set_title('우리반 영어,수학,국어 점수 차트')

plt.show()
```

```python
# box plot 그리기
plt.style.use('ggplot')
fig = plt.figure()

ax= fig.add_subplot(111)
x = [1,2,3,3,11,20]
ax.boxplot(x)
plt.show()
```

```python
# 여러개의 box plot 만들기
plt.style.use('ggplot')
fig = plt.figure()

ax= fig.add_subplot(111)
label = ['english','math']

x = [[1,2,3,3,11,20],[1,2,9,10,19,13]]
ax.boxplot(x,labels=label)
plt.show()
```
