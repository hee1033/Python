```python
# requests 모듈 -> 웹페이지를 요청하고 응답데이터를 받는 모듈
import requests as rq
from bs4 import BeautifulSoup
# 특정 사이트 페이지 요청
response = rq.get("https://news.naver.com/")
print(response.text)
```

```python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
print(html_doc)
```

```python
soup = BeautifulSoup(html_doc, 'html.parser') # soup을 이용해서 html_doc 문자열을 html 파싱
list1 = soup.select("a") # 선택지를 이용해 태그선택, 결과는 리스트로 반환
print(list1[1])
print(list1[1].text) # 텍스트를 가져오기
```

```python
# 속성값을 가져오기 <태그명 속성명="속성값">
print(list1[1]["href"])
```
